## Reading the file with Pandas,
#### 'Feature ID' is used as an index for the file

```python
findice = pd.read_csv('../../panopea/panopea_data/data-results/Male-v-Female-Expression.csv', index_col=0) # abre con indice
findice
```

This script obtains the number of sequences that are:

a) unique for females (Female-Unique)

b) unique for males (Male-Unique)

c) male reads (Male-Unique) with an expression level higher than 100 than the females (Female-Unique)

d) male reads (Male-Unique) with an expression level lower than 0.01 than the females (Female-Unique)

expression level is measured as 'Male-Unique']/'Female-Unique' larger than 100 or lower than 0.001
Data are saved in a new dataframe, which could be later saved as a csv file.

```python
mafa=[]
linea=''
n=200
for rows in findice.index:
    row= findice.ix[rows]
    linea=''
    if row['Male-Unique']==0 and row['Female-Unique']>n:
        linea= rows,str(row['Male-Unique']),str(row['Female-Unique']),"female" 
    elif row['Male-Unique']>n and row['Female-Unique']==0:
        linea= rows,str(row['Male-Unique']),str(row['Female-Unique']), "male"
    elif row['Male-Unique']!=0 and row['Female-Unique']!=0 :#and row['Male-Unique']>n and row['Female-Unique']>n:
        if row['Male-Unique']/row['Female-Unique']>100:
            linea= rows,str(row['Male-Unique']),str(row['Female-Unique']), "male-female"
        elif row['Male-Unique']/row['Female-Unique']<.01:
            linea= rows,str(row['Male-Unique']),str(row['Female-Unique']), "female-male"
    if linea !='':
        #print (linea)
        mafa.append(linea)

mafa= pd.DataFrame(mafa, columns=['Feature ID', 'Male-Unique', 'Female-Unique', 'sex'])

mafa.describe()
```

```python
# if you want to see the data base, run this command
mafa
```

Grouping data by

a) female only

b) male only

c) female-male: contigs with a higher expression in female

d) male-female: contigs with a higher expression in male

```python
mafa2=mafa.groupby('sex')
mafa2.describe()
```

### Saving file

```python
mafa.to_csv('../../panopea/panopea_data/data-results/mafa.csv')
```

# If you want to obtain the differential expression with Total 
# reads and differential values < 0.01 or > 100 ratio

```python
mafa1=[]
linea=''
n=200
for rows in findice.index:
    row= findice.ix[rows]
    linea=''
    if row['Male-Unique']==0 and row['Female-Unique']>n:
        linea= rows,str(row['Male-Total']),str(row['Female-Total']),"female" 
    elif row['Male-Unique']>n and row['Female-Unique']==0:
        linea= rows,str(row['Male-Total']),str(row['Female-Total']), "male"
    elif row['Male-Total']!=0 and row['Female-Total']!=0 :#and row['Male-Total']>n and row['Female-Total']>n:
        if row['Male-Total']/row['Female-Total']>100:
            linea= rows,str(row['Male-Total']),str(row['Female-Total']), "male-female"
        elif row['Male-Total']/row['Female-Total']<.01:
            linea= rows,str(row['Male-Total']),str(row['Female-Total']), "female-male"
    if linea !='':
        #print (linea)
        mafa1.append(linea)

mafa1= pd.DataFrame(mafa1, columns=['Feature ID', 'Male-Total', 'Female-Total', 'sex'])
mafa1.describe()
```

```python
mafa2=mafa1.groupby('sex')
mafa2.describe()
```
