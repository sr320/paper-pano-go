```python
bigmicrosin = pd.merge(bigv8, micros, on= "ContigID", how="inner")
bigmicrosin
```

## grouping for motif of the microsatellites

```python
mp = bigmicrosin.groupby("motif")["ContigID"].count()
mp
```

## to obtain those annotated contigs with microsatellites

```python
df = bigmicrosin[bigmicrosin["ProteinName"].notnull()]
df
```

## there are 130 annotated microsatellites

## grouping by motif of the microsatellite

```python
mp = df.groupby("motif")["ContigID"].count()
mp
```

## grouping by sex-based expression

```python
mp2 = df.groupby("sex-based_expresssion")["ContigID"].count()
mp2
```

### there are only three microsatellites associated with over-expressed genes in females

```python
for row in df.index:
    row1 = df.ix[row]
    if str(row1["sex-based_expresssion"])!="nan":
        print (row1["ContigID"], row1["sex-based_expresssion"] , str(row1["ProteinName"])[:60])
```
