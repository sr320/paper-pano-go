## criteria for designating "male" or "female"
1. both 'female_unique_count' and 'male_unique_count' most have reads
2. if one of them is zero the other most to have more than 200 reads
 1. to asign "male", 'female_unique_count'=0, 'male_unique_count'>200
 2. to asign "female", 'male_unique_count'=0, 'female_unique_count'>200
3. to asign "male", the ratio 'male_unique_count'/'female_unique_count' most be > 100
4. to asign "female", the ratio 'male_unique_count'/'female_unique_count' most be < 0.01

````python
mafa=[]
linea=''
n=0
n1=0
n2=0
#print ('ContigID        \t ratio  \t sex-based   \t sb \t expr \t male \t female' )
for rows in f.index:
    row= f.ix[rows]
    linea=''
    espacio = ''
    if row['female_unique_count']==0 and row['male_unique_count']==0: # both are zero!
        linea= row['ContigID'],"", 0.0
        espacio= ' '
    elif row['female_unique_count']>200 and row['male_unique_count']==0: 
        # female >200 reads, male = 0
        linea= row['ContigID'],"female" , (row['female_unique_count']+0.0)
    elif row['male_unique_count']>200 and row['female_unique_count']==0: 
        # male >200 reads, female = 0
        linea= row['ContigID'],"male" ,(row['male_unique_count']+0.0)
    elif row['male_unique_count']/row['female_unique_count']>100 and row['female_unique_count']!= 0: 
        # ratio male/female > 100 (100 times number of reads in males) 
        linea= row['ContigID'],"male" , round(row['male_unique_count']/row['female_unique_count'],4)
    elif row['male_unique_count']/row['female_unique_count']<0.01 and row['male_unique_count']!= 0:
        linea= row['ContigID'],"female" , round(row['male_unique_count']/row['female_unique_count'],4)
        # ratio male/female < 0.001 (100 times number of reads in females) 
    else:
        linea= row['ContigID'],"", 0.0
        espacio= ' '
    mafa.append(linea)
    n+=1

mafa= pd.DataFrame(mafa, columns=['ContigID', 'sex', 'ratio'])
print (n, "sequences ")
````

````python
fprimers = pd.read_csv ("../panopea_data/data-results/geoduck-v3-msat/geoduck-v3primers.csv", index_col=None, 
                 engine = "python")
fprimers
````

````python
# eliminating contigs with duplicate primers
fprimers_unique = fprimers[fprimers.duplicate==0]
fprimers_unique
````

````python
# change name to column "name" to "ContigID" and adding the column "primers"
fprimers_unique =fprimers_unique.rename(columns={"name": 'ContigID'}) 
fprimers_unique["primers"] = "yes"
fprimers_unique
````

````python
#checking for duplicated contigs
fprimers_unique = fprimers_unique.drop_duplicates(subset=("ContigID"), inplace = False)
fprimers_unique
````

````python
# extracting useful columns to be added to main file
df1 = pd.DataFrame(fprimers_unique["ContigID"], columns= [ "ContigID"])
df2 = pd.DataFrame(fprimers_unique["primers"], columns= ["primers"])
df3 = pd.concat([df1, df2], axis=1)
df3
````

````python
f2 = pd.merge(f2,df3, how='outer')
f2
````

# how many microsatellites have an annotation?

````python
n=0
lista_m_a =[]
print ("Contigs with annotated microsatellites")
for rows in f2.index:
    row = f2.ix[rows]
    if row["primers"]=="yes" and str(row["ProteinName"])!="nan":
        n+=1
        print (n, "\t", row["ContigID"], "\t", row["ProteinName"][:60])
        lista_m_a.append(row["ContigID"])

print ("\n", n, "\t", "microsatellites with primers and annotation")
````
