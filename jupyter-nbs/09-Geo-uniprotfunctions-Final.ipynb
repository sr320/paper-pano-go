### Reading the Uniprot_sprot file
It may take a while (~90 s)

```python
uniprot = SeqIO.index("../../bigdata/uniprot_sprot.dat", "swiss")
```

### Reading file for Males

```python
findice = pd.read_csv('../../panopea/panopea_data/data-results/male_protein_names.csv', index_col=0)
findice
```

### Extracting function and saving file

```python
secuencias=[]
sec=''
linea=''
for acc in findice['UniProt_Acc']:
    sec =str(uniprot.get_raw(acc))
    sec1=sec[sec.find("FUNCTION:"):]    # finds the function of the protein
    sec1=sec1.replace("FUNCTION: ","")  # eliminates the word "FUNCTION: "
    sec1=sec1[:sec1.find("-!-")]        # finds the end of the function
    sec1=sec1.replace("\\n","")         # eliminates "line-return" 
    sec1=sec1.replace("CC"," ")         # eliminates "CC" caracters used as continuation
    sec1=sec1.replace("   "," ")        # elimintas extra spaces
    sec1=sec1.replace("  "," ")         # idem
    #print ( acc+ "\t" + sec1)           # prints on screen eliminate initial # to print on screen
    #print()
    linea=acc,sec1
    secuencias.append(linea)

secuencias= pd.DataFrame(secuencias, columns=['UniProt_Acc', 'function'])
secuencias
```

```python
secuencias.to_csv('../../panopea/panopea_data/data-results/male_protein_functions.csv', index_col=0)
```

```python
secuencias=pd.read_csv('../../panopea/panopea_data/data-results/male_protein_functions.csv', index_col=0)
secuencias
```

### merging files and saving
